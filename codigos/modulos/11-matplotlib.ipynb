## Alterando informações da visualização

- Para isso é importante importar a biblioteca matplotlib

```python
import matplotlib.pyplot as plt
```

```python
df = arquivo['STATE'].value_counts(ascending=True).tail()
dr = arquivo.groupby(arquivo['STATE'])['ANNUAL'].mean().sort_values(ascending=False).head(4)
```

#### title='nome do título' - Altera o título do gráfico

```python
df.plot.bar(title='Quantitdade de cidades por estado');
```

```python
df.plot.pie(title='Estados com mais cidades');
```

#### .xlabel('Nome') - Altera o nome do eixo x
#### .ylabel('Nome') - Altera o nome do eixo y

```python
df.plot.bar(title='Estados com mais cidades')
plt.xlabel('Estado')
plt.ylabel('Quantidade cidades');
```

```python
df.plot.pie(title='Estados com mais cidades')
plt.ylabel('');
```

#### color='cor em inglês' - Altera a cor do gráfico

```python
df.plot.bar(title='Estados com mais cidades', color='black')
plt.xlabel('Estado')
plt.ylabel('Quantidade cidades');
```

#### color=[lista de cores] - Muda mais de uma cor

```python
df.plot.bar(title='Estados com mais cidades', color=['black', 'gray'])
plt.xlabel('Estado')
plt.ylabel('Quantidade cidades');
```

#### .style.available - Retorna os estilos

```python
print(plt.style.available)
```

#### .style.use('estilo') - Muda o estilo do gráfico

- Todos os estilos estão no site do matplotlib
- O estilo é aplicado globalmente nos gráficos
- https://matplotlib.org/

```python
plt.style.use('classic')
df.plot.bar()
plt.show; # Mostra o gráfico
```

#### Utilizando estilos sem ser globalmente
with plt.style.context('estilo'):

    arquivo.plot()

```python
with plt.style.context('seaborn-deep'):
    dr.plot.bar()
```

#### .legend() - Mostra a legenda do gráfico

```python
with plt.style.context(['classic','dark_background']):
    dr.plot.bar(color=['green', 'yellow', 'blue', 'white'], title='Estados com maior produção')
plt.legend()
plt.show;
```

```python
plt.style.use('_mpl-gallery-nogrid')
```

#### marker='caractere' - Acrescenta marcas nos pontos do gráfico

```python
with plt.style.context(['classic','ggplot']):
    dr.plot(marker='o', color='y')
```

## Salvando em png

#### .savefig("nome_do_arquivo.png") - Salva o gráfico como png

```python
with plt.style.context(['classic','ggplot']):
    dr.plot(marker='o', color='m')
plt.legend()
plt.xlabel('Média anual')
plt.ylabel('Estado')
plt.savefig('grafico.png')
```
