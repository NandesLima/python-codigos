# Pandas

- Biblioteca para análise de dados.

- Dá ao Python a capacidade de trabalhar com dados do tipo planilha.

- Permite carregar, manipular, alinhar e combinar dados rapidamente, entre outras funções.

```python
import pandas
```

```python
arquivo = "dataset.csv"
```

### Leitura de arquivos .csv

#### .read_csv() - Recebe como parâmetro o arquivo que será lido

```python
df = pandas.read_csv(arquivo)

print(df)
```

#### error_bad_lines=False - Parâmetro que faz a leitura ignorar linhas com erro

```python
df = pandas.read_csv(arquivo, error_bad_lines=False)
print(df)
```

#### sep="," - Parâmetro que troca o caractere de separação, por padrão é a vírgula

```python
df = pandas.read_csv(arquivo, sep=";")
print(df)
```

#### .groupby() - Faz o agrupamento de dados

- Escolhe por qual coluna quer fazer o agrupamento, dentro do parênteses.
- Utiliza-se um colchetes com o parÂmetro que será utilizado como filtro.
- Usa um método que vai gerar a interação desejada

```python
# Contagem das cidades de cada estado
df.groupby('Estado')['Cidade'].nunique()
```

```python
# Média de irradiação solar anual por estado
df.groupby("Estado")["Anual"].mean()
```

```python
# Desvio padrão da irradiação solar anual por estado
df.groupby("Estado")["Anual"].std()
```

#### .isnull() - Verifica se existe valores nulos

```python
nordeste.isnull()
```

#### .isnull().sum() - Verifica a quantidade de nulos

```python
nordeste.isnull().sum()
```

#### .fillna() - Recebe como parâmetro um valor que será colocado em espaços nulos

- inplace=True, parâmetro que substitui o objeto em memória, ou seja salvas modificações

```python
nordeste["ANNUAL"].fillna(nordeste["ANNUAL"].mean(), inplace=True)

nordeste.isnull().sum()
```

```python
nordeste["DEC"].fillna(5700, inplace=True)

nordeste.isnull().sum()
```

#### .dropna() - Remove linhas com valores faltantes

- inplace=True, parâmetro que substitui o objeto em memória, ou seja salvas modificações

- subset=[coluna], parâmetro para apagar linhas com valores nulos com base em uma coluna

- how='all', parâmetro que remove linhas que todas suas colunas estejam como nulas

```python
nordeste.dropna(subset=['NOV'], inplace=True)

nordeste.isnull().sum()
```

```python
nordeste.dropna(how='all', inplace=True)

nordeste.isnull().sum()
```

#### .nlargest() - Utiliza dois parâmetros, o primeiro é a quantidade de linhas que ele deve retornar, o segundo é a coluna que vai ser analisada. No final é retornado os que tiverem os maiores valores da coluna indicada.

```python
df.nlargest(5, "Anual")
```

#### .nsmallest() - Utiliza dois parâmetros, o primeiro é a quantidade de linhas que ele deve retornar, o segundo é a coluna que vai ser analisada. No final é retornado os que tiverem os menores valores da coluna indicada.

```python
df.nsmallest(7, "FEB")
```

#### .sort_values() - Retorna o conjunto de dados ordenado pela coluna escolhida

```python
df.sort_values("Anual")
```

```python
df.sort_values("Anual").head(5)
```

#### ascending=False, parâmetro que faz com que a ordem seja do menor para o maior

```python
df.sort_values("Anual", ascending=False).head(10)
```

## Criando colunas

- Para criar novas colunas, precisa-se apenas declarar a variável com um nome de coluna ainda não utilizado e atribuir valores

```python
nordeste['Total_anual'] = nordeste['JAN'] + nordeste['FEB'] + nordeste['MAR'] + nordeste['APR'] + nordeste['MAY'] + nordeste['JUN'] + nordeste['JUL'] + nordeste['AUG'] + nordeste['SEP'] + nordeste['OCT'] + nordeste['NOV'] + nordeste['DEC']

nordeste.head()
```

```python
nordeste['Media_S1'] = (nordeste['JAN'] + nordeste['FEB'] + nordeste['MAR'] + nordeste['APR'] + nordeste['MAY'] + nordeste['JUN'])/6

nordeste.head()
```

```python
nordeste['Media_S2'] = (nordeste['JUL'] + nordeste['AUG'] + nordeste['SEP'] + nordeste['OCT'] + nordeste['NOV'] + nordeste['DEC'])/6

nordeste.head()
```

```python
nordeste.groupby("STATE")["ANNUAL"].std()
```
