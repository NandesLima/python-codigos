# 03 - Retorno de escopo

## dir() - Retorna o escopo local ou de um objeto específico em uma lista

```python
dir()
```

```python
dir(int)
```

```python
dir(dir)
```

```python
dir(25.3)
```

## help() - Retorna o escopo, método, classes e outras informações. Do local ou objeto especificado. Serve para conhecer programas.

```python
help()
```

```python
help(bin)
```

```python
help(oct)
```
